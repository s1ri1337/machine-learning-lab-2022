```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer

from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('HousePrice.csv')
df
```

```python
# dropping irrelevant columns with null values
values = ['Id', 'Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature']
df.drop(values, axis=1, inplace=True)
df
```

## Q2 Split the dataset into train size of 70% and test size of 30% and Apply the Ridge and Lasso regression and fit the model containing all independent variables.

```python
x = df.copy()
y = x.pop('SalePrice')

x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=742)
```

```python
#splitting Numerical and Categorical Columns

#for numerical columns
numerical_cols   = [c for c in x.columns if x[c].dtype in ['int64', 'float64']]\
#for categorical columns
categorical_cols = [c for c in x.columns if x[c].dtype == 'object']
```

```python
#finding number of unique values in categorical columns
cat_att = []
for i in df.columns:
    if df[i].dtype == 'O':
        print(i)
        cat_att.append(i)
        print(f"Number of unique values in column {i}:", df[i].nunique())
```

```python
#creating pipeline

#scaling the numerical columns
num_imputer = SimpleImputer(strategy='mean')
cat_imputer = SimpleImputer(strategy='median')

numerical_transformer = Pipeline(steps = [
    ('scaler', RobustScaler()),
    ('imputer', num_imputer)
])

#one-hot encoding the categorical columns
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore')),
    ('imputer', cat_imputer)
])

#bundle preprocessing for numerical and categorical data
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols),
      ])
```

```python
lasso = Lasso(alpha=0.1)

pipe_l = Pipeline(steps=[('preprocessor', preprocessor), 
                              ('model', lasso)])
```

```python
pipe_l.fit(x_train, y_train)
```

```python
#lasso model accuracy
pipe_l.score(x_test, y_test)
```

```python
ridge = Ridge(alpha=1.0)

pipe_r = Pipeline(steps=[('preprocessor', preprocessor), 
                              ('model', ridge)])
```

```python
pipe_r.fit(x_train, y_train)
```

```python
#ridge model accuracy
pipe_r.score(x_test, y_test)
```
