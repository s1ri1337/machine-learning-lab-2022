### 2. Split the dataset into train and test with 80:20 split and Bulid the linear regression model. 

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('SOLD PRICE', axis = 1)
y = df.pop('SOLD PRICE')

train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.2, random_state = 42) #dataset is split into train and test
```

```python
numerical_cols   = [c for c in x.columns if x[c].dtype in ['int64', 'float64']]
categorical_cols = [c for c in x.columns if x[c].dtype == 'object']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
```

```python
num_imputer = SimpleImputer(strategy='mean')
cat_imputer = SimpleImputer(strategy='median')

numerical_transformer = Pipeline(steps = [
    ('scaler', RobustScaler()),
    ('imputer', num_imputer)
])

categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore')),
    ('imputer', cat_imputer)
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols),
      ])
```

```python
linreg = LinearRegression()
pipe_lr = Pipeline(steps=[('preprocessor', preprocessor), 
                              ('model', linreg)])
```

```python
pipe_lr.fit(train_x, train_y)
```

```python
pipe_lr.score(test_x, test_y)
```

```python
prediction_y_lr = pipe_lr.predict(test_x)
```

### 3. Apply the Stochastic Gradient Descent to minimize the cost function

```python
from sklearn.linear_model import SGDRegressor
```

```python
sgd = SGDRegressor()
pipe_sgd_lr = Pipeline(steps=[('preprocessor', preprocessor), 
                              ('model', sgd)])
pipe_sgd_lr.fit(train_x, train_y)
```

```python
pipe_sgd_lr.score(test_x, test_y)
```

```python
prediction_y_sgd_lr = pipe_sgd_lr.predict(test_x)
```

#### We can see there is a slight increase in the score compared to normal linear regression

### 4. Check whether the model is overfitting. Apply the L2 regulerization if the model overfit.

#### No, the model is not overfitting here. There is no need for regularization

### 5. Tabulate and write down the relevant measures of accuracy for both models.

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
print("Linear Regression")
r2s = r2_score(test_y,prediction_y_lr)
mse = mean_squared_error(test_y,prediction_y_lr)
print('R2 score =',r2s)
print('Mean Squared Error =', mse)
print('Root Mean Squared Error =', np.sqrt(mse))

print("\nLinear Regression with stochastic gradient descent")
r2s = r2_score(test_y,prediction_y_sgd_lr)
mse = mean_squared_error(test_y,prediction_y_sgd_lr)
print('R2 score =',r2s)
print('Mean Squared Error =', mse)
print('Root Mean Squared Error =', np.sqrt(mse))
```
